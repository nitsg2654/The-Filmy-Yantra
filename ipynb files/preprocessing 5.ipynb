```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('blwd_movies.csv')
```

```python
data.head(10)
```

```python
data = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
data
```

```python
data['actor_1_name'] = data['actor_1_name'].replace(np.nan, 'unknown')
data['actor_2_name'] = data['actor_2_name'].replace(np.nan, 'unknown')
data['actor_3_name'] = data['actor_3_name'].replace(np.nan, 'unknown')
data['director_name'] = data['director_name'].replace(np.nan, 'unknown')
```

```python
data['genres'] = data['genres'].str.replace(',', '')
```

```python
data
```

```python
data['movie_title'] = data['movie_title'].str.lower()
```

```python
data['movie_title'][1]
```

```python
data.isna().sum()
```

```python
data['comb'] = data['actor_1_name'] + ' ' + data['actor_2_name'] + ' '+ data['actor_3_name'] + ' '+ data['director_name'] +' ' + data['genres']
```

```python
data
```

```python
old_data = pd.read_csv("main_data.csv")
old_data
```

```python
new_data = old_data._append(data,ignore_index=True)
new_data
```

```python
new_data.to_csv("main_data.csv", index=False)
```
